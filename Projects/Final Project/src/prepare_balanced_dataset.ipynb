```python
df['gender'] = df['gender'].str.strip().str.lower()
```

```python
df = df[df['gender'].isin(values=['male', 'female'])]
```

```python
df.info()
```

```python
df['id'].value_counts()
```

```python
features_cols = [col for col in df.columns if col not in ['id', 'gender', 'file_name']]
```

```python
df_sub = df.groupby(['id', 'gender'])[features_cols].mean().reset_index()
```

```python
df_sub
```

```python
df_sub['gender'].value_counts()
```

```python
gender_series = df_sub['gender'] == 'male'
gender_series
```

```python
index_male = list(gender_series[gender_series == True].index)
```

```python
gender_series
```

```python
selected_index_male = np.random.choice(index_male, size=28, replace=False)
```

```python
selected_index_male
```

```python
df_woman = df_sub[df_sub['gender'] == 'female']
```

```python
df_woman.head()
```

```python
df_sub
```

```python
df_man = df_sub.iloc[selected_index_male].reset_index(drop=True)
```

```python
df_combined = pd.concat([df_man, df_woman], axis=0).reset_index(drop=True)
```
